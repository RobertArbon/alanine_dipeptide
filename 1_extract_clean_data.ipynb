```python
import pandas as pd
import numpy as np
%matplotlib inline
import re
```

```python
trial_path = 'data/ala1_trials.csv'
df = pd.read_csv(trial_path)
df.T
```

## Summary statistics

```python
def f(x, **kwargs):
    try: 
        y = re.findall('([0-9]\.[0-9]*)\,', str(x))
        y = [float(i) for i in y]
        g = kwargs.pop('stat')
        result = g(y, **kwargs)
    except:
        result = x
        
    return result
    
for col in ['test', 'train']:
    df.loc[:, col+'_mean'] = df[col+'_scores'].apply(f, stat=np.mean)
    df.loc[:, col+'_std'] =  df[col+'_scores'].apply(f, stat=np.std)
    df.loc[:, col+'_min'] =  df[col+'_scores'].apply(f, stat=np.min)
    df.loc[:, col+'_max'] =  df[col+'_scores'].apply(f, stat=np.max)
    df.loc[:, col+'_q25'] =  df[col+'_scores'].apply(f, stat=np.quantile, q=0.25)
    df.loc[:, col+'_q75'] =  df[col+'_scores'].apply(f, stat=np.quantile, q=0.75)
```

```python
df.T
```

## Parameters

```python
def get_dict(x):
    return dict(re.findall("'(.*?)'\: (.*?)[,\}]", x))

df.loc[:, 'parameters'] = df['parameters'].apply(get_dict)
```

```python
keys = list(df['parameters'][0].keys())
for k in keys: 
    df.loc[:, k] = df['parameters'].apply(lambda x: x[k])
```

```python
for k in keys:
    try:
        df.loc[:, k] = df.loc[:, k].astype(float)
    except ValueError: 
        df.loc[:, k] = df.loc[:, k].astype("category")
```

## Drop redundant columns
